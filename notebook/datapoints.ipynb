```python
import pandas as pd
import numpy as np
```

```python
df1= pd.read_excel("../data/External_Cibil_Dataset.xlsx")
df2= pd.read_excel("../data/Internal_Bank_Dataset.xlsx")
```

```python
df2.shape
```

```python
External_unknown = df1.sample(n=500, random_state=42)
```

```python
unknown_ids = External_unknown["PROSPECTID"]
```

```python
External_training = df1[~df1["PROSPECTID"].isin(unknown_ids)]
```

```python
Internal_unknown = df2[df2["PROSPECTID"].isin(unknown_ids)]

Internal_training = df2[~df2["PROSPECTID"].isin(unknown_ids)]
```

```python
Internal_unknown.shape
```

```python
External_unknown = External_unknown.drop(columns=["Approved_Flag"])
```

```python
External_unknown.to_excel("../data/new_external_data.xlsx", index=False)
Internal_unknown.to_excel("../data/new_internal_data.xlsx", index=False)

External_training.to_excel("../data/train_external_data.xlsx", index=False)
Internal_training.to_excel("../data/train_internal_data.xlsx", index=False)
```
